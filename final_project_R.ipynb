```r
#import data
kmsn <- read.csv("kmsn_2.csv",header =TRUE,fileEncoding = 'UTF-8-BOM')

#Export columns
head(kmsn)
```

```r
# NOW WE WANT TO CONSIDER MONTHLY EXTREME EVENTS
# Develop a time series of monthly maxima
year<-kmsn$year
month<-kmsn$month
prcp<- kmsn$PRCP

# Look at your mydata vector and you can see the first year and the last year on record
startYear = 1955
endYear = 2018
allyears<-c(startYear:endYear)
```

```r
newval<-cbind(year,prcp)
yearMax<- aggregate(prcp ~ year, newval,max) 
```

```r
#Allocate empty arrays to fill in your loop below
monthMax<-c()
timeYear<-c()
timeMonth<-c()
timeTobs<-c()
timeSOI<-c()
timePNA<-c()
timeMEI<-c()
timePDO<-c()
timeNAO<-c()
cnt = 1 #designates where to fill in your empty arrays, increments by one

# The loop below loops through each year and month on record and saves the monthly maximum value each year
for (i in 1:length(allyears)) {
  for (k in 1:12) {
    
    ind<-which(year %in% allyears[i])
    ind2<-which(month[ind] %in% k)
    monthMax[cnt] <- max(prcp[ind[ind2]]) #highlights data by column
    timeYear[cnt] <- allyears[i] #highlights data by column
    timeMonth[cnt] <- k #highlights data by column
    timeTobs[cnt] <- kmsn$doy[ind[ind2]] #highlights data by column
    timeSOI[cnt] <- kmsn$SOI[ind[ind2]]
    timeMEI[cnt] <- kmsn$MEI[ind[ind2]]
    timePDO[cnt] <- kmsn$PDO[ind[ind2]]
    timeNAO[cnt] <- kmsn$NAO[ind[ind2]]
    
    
    cnt = cnt+1 #increment by one
  }
  
}


```

```r
##Go through each row and determine if a value is zero
row_sub = apply(kmsn[5], 1, function(row) all(row !=0 ))
##Subset as usual
kmsn_0=kmsn[row_sub,]
```

```r
# set 99% threshold
u<-quantile(kmsn_0$PRCP, probs=c(.99))
u
```

```r
#Use the intervals method to decluster your data
extremalindex(kmsn_0$PRCP, u, method = c("intervals"))
declust_x<-decluster(kmsn_0$PRCP,u,method=c("intervals"))

plot(declust_x)
```

```r
# Plot monthly data
plot(kmsn$month,kmsn$PRCP,xlab="month",ylab="precipitation (in)",col="black")
```

```r
# First we allocate our arrays
monthlyScale = c()
monthlyShape = c()
monthlyLocation = c()
monthlyReturnLevel_High= c() 
monthlyReturnLevel_Best = c()
monthlyReturnLevel_Low = c()

# Run the loop for months 1 through 12
for (k in 1:12) {
  
  ind<-which(timeMonth %in% k) #find location in our time series that is the correct month
  fit <- fevd(monthMax[ind]) # Fit our GEV distribution
  
  # SAVE THOSE FITS FOR EACH MONTH
  monthlyLocation[k]<-fit$results$par[1]
  monthlyScale[k]<-fit$results$par[2]
  monthlyShape[k]<-fit$results$par[3]
  #DO THE SAME FOR THE RETURN PERIOD
  rp<-return.level(fit,return.period = c(100),do.ci=TRUE)
  monthlyReturnLevel_High[k]<-rp[3] #UPPER CONFIDENCE BOUND
  monthlyReturnLevel_Low[k]<-rp[1] #LOWER CONFIDENCE BOUND
  monthlyReturnLevel_Best[k]<-rp[2]   #BEST ESTIMATE
  
}
```

```r
# PLOT THE PARAMETER ESTIMATES -- CAN SEE ANNUAL SIGNAL IN PARAMETERS
x0<-c(1:12)

# 3 figures arranged in 3 rows and 1 column
par(mfrow=c(3,1))
plot(x0,monthlyLocation,ylab='location',xlab='month', pch=16, cex=2)
plot(x0,monthlyScale,ylab='scale',xlab='month', pch=16, cex=2)
plot(x0,monthlyShape,ylab='shape',xlab='month', pch=16, cex=2)
```

```r
#stationary fit
fit0<-fevd(monthMax)
fit0
```

```r
# NON-STAT FIT 1: ANNUAL HARMONIC in LOCATION PARAMETER
fit1 <- fevd(monthMax,location.fun =
               ~ cos(2 * pi * timeTobs / 365.25) + sin(2 * pi * timeTobs / 365.25)) #ANNUAL HARMONIC

fit1
```

```r
# NON-STAT FIT 2: ANNUAL HARMONIC SCALE PARAMETER
fit2<- fevd(monthMax,
           scale.fun=~ cos(2 * pi * timeTobs / 365.25) + sin(2 * pi * timeTobs / 365.25)) #ANNUAL HARMONIC

fit2
```

```r
# NON-STAT FIT 3: ANNUAL HARMONIC in LOCATION AND SCALE PARAMETER
fit3<- fevd(monthMax,location.fun =
               ~ cos(2 * pi * timeTobs / 365.25) + sin(2 * pi * timeTobs / 365.25),
           scale.fun=~ cos(2 * pi * timeTobs / 365.25) + sin(2 * pi * timeTobs / 365.25),type=c("GEV")) #ANNUAL HARMONIC

fit3
```

```r
#lr tests for stationary GEV vs non-stationoaries 
lr.test(fit0, fit1)
lr.test(fit0, fit2)
lr.test(fit0, fit3)
```

```r
# set climate indicies
SOI<-kmsn$SOI
MEI<-kmsn$MEI
PDO<-kmsn$PDO

# fit covariates
fitSOI <- fevd(monthMax, location.fun =  ~timeSOI)
fitMEI <- fevd(monthMax, location.fun =  ~timeMEI)
fitPDO <- fevd(monthMax, location.fun =  ~timePDO)

fitSOI
fitMEI 
fitPDO
```

```r
# parameter estimates for fit3 (best fit)
mu0<-fit3$results$par[1]
mu1<-fit3$results$par[2]
mu2<-fit3$results$par[3]
sigma0<-fit3$results$par[4]
sigma1<-fit3$results$par[5]
sigma2<-fit3$results$par[6]
xi<-fit3$results$par[7]

#COMPUTE YOUR TIME VARYING LOCATION PARAMETER, here we are only doing it for the 
#first day representing each month, hence 
#timeTobs[1:12] #Gives you the first day of each month
mu<- mu0+mu1*(cos(2 * pi * timeTobs[1:12] / 365.25)) +mu2*( sin(2 * pi * timeTobs[1:12] / 365.25))
sig<- sigma0+sigma1*(cos(2 * pi * timeTobs[1:12] / 365.25))+sigma2*(sin(2 * pi * timeTobs[1:12] / 365.25))

# CAN GET THE "BEST" RETURN LEVEL ESTIMATE TWO WAYS
# METHOD 1 - PLUG YOUR ESTIMATES INTO THE RETURN LEVEL FORMULA
p = 1/100 #20 year event
RL_nonstat<-mu-(sig/xi)*(1-(-log(1-p))^(-xi))
RL_nonstat
monthlyReturnLevel_Best
# You don't get CI this way though.

# METHOD 2 - PLUG INTO YOU RETURN LEVEL FUNCTION BY SETTING UP A NUMERIC MATRIX FOR COVARIATES
v <- make.qcov(fit3, vals = list(
  mu1 = cos(2 * pi * timeTobs[1:12] / 365.25), mu2 = sin(2 * pi * timeTobs[1:12] / 365.25), 
  sigma1 = cos(2 * pi * timeTobs[1:12] / 365.25), sigma2 = sin(2 * pi * timeTobs[1:12] / 365.25)))

#Look at first 10 lines of your output
v[1:10, ]

RL_all<-return.level(fit3, return.period = c(100), qcov = v,do.ci=TRUE)
RL <-RL_all[,2]
RL_u<-RL_all[,3] #upper bound
RL_l<-RL_all[,1] # lower bound


# RL and RL NONSTAT SHOULD BE THE SAME!!

# Now we can compare our plots!!

# 3 figures arranged in 3 rows and 1 column
par(mfrow=c(1,1))
plot(x0,  monthlyReturnLevel_Best, ylim=c(0,8), ylab="100-yr return level (in)", xlab="month", pch=16, cex=2)
points(timeMonth,monthMax,col="blue")

arrows(x0=x0, y0=monthlyReturnLevel_Low, x1=x0, y1=monthlyReturnLevel_High, code=3, angle=90, length=0.1)

lines(RL,col="black",lwd=2)
lines(RL_u,lty=2)
lines(RL_l,lty=2)
#abline(h=5.5,col="green")

legend("bottomleft",c("monthly GEV","Non-Stat GEV","Non-stat GEV 95% CI"),
       col=c("blue","black","black", "black"),pch=c(16,16,NA,NA), lty=c(NA,NA,1,2), cex=0.8)

title("monthly 100-year return levels")

RL_u
RL_l
```
